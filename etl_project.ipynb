```python
# Create Engine and Pass in MySQL Connection to create a database
db_engine = create_engine("mysql://root:<amigo2598>@localhost")
```

### Create the database SCHEMA crimes_db

```python
# create the database that will be used 
db_engine.execute("CREATE DATABASE if not exists crime_db")            
```

### Create the us_states table 

```python
# create the us state table
db_engine.execute("USE crime_db")
db_engine.execute("CREATE TABLE us_states( \
                        id INT AUTO_INCREMENT NOT NULL, \
                        fips_state_code INT NOT NULL, \
                        state_code VARCHAR(2) NOT NULL, \
                        state_name VARCHAR(60) NOT NULL, \
                        PRIMARY KEY (id))")
```

### Create the us_counties table 

```python
# create the us county table
db_engine.execute("USE crime_db")
db_engine.execute("CREATE TABLE us_counties( \
                        id INT AUTO_INCREMENT NOT NULL, \
                        state_code VARCHAR(2) NOT NULL, \
                        fips_state_code INT NOT NULL, \
                        fips_county_code INT NOT NULL, \
                        county_name VARCHAR(60) NOT NULL, \
                        PRIMARY KEY (id))")
```

### Create the us_crime table

```python
# create the us crime table
db_engine.execute("USE crime_db")
db_engine.execute("CREATE TABLE us_crimes( \
                        id INT AUTO_INCREMENT NOT NULL, \
                        fips_state_code INT NOT NULL, \
                        fips_county_code INT NOT NULL, \
                        crime_rate INT NOT NULL, \
                        crime_murder_count INT NOT NULL, \
                        crime_rape_count INT NOT NULL, \
                        crime_robbery_count INT NOT NULL, \
                        crime_agasslt_count INT NOT NULL, \
                        crime_burglry_count INT NOT NULL, \
                        crime_larceny_count INT NOT NULL, \
                        crime_mvtheft_count INT NOT NULL,\
                        crime_arson_count INT NOT NULL, \
                        county_population INT NOT NULL, \
                        PRIMARY KEY (id))")
```

### Create the us_education table

```python
# create the us crime table
db_engine.execute("USE crime_db")
db_engine.execute("CREATE TABLE us_education( \
                        id INT AUTO_INCREMENT NOT NULL, \
                        fips_state_code INT NOT NULL, \
                        fips_county_code INT NOT NULL, \
                        hs_diploma_y_count INT NOT NULL, \
                        hs_diploma_n_count INT NOT NULL, \
                        as_degree_count INT NOT NULL, \
                        ba_degree_count INT NOT NULL, \
                        PRIMARY KEY (id))")
```

# Store US States CSV into DataFrame

```python
# read the us state csv into a dataframe and rename the columns
csv_file = "Resources/us_states.csv"
us_states_data_df = pd.read_csv(csv_file)
us_states_data_df.rename(columns={"fips_code":"fips_state_code",
                                  "post_code":"state_code",
                                  "state":"state_name"},inplace=True)
us_states_data_df.head()
```

```python
# validate that the data was loaded into the table
pd.read_sql_query('select * from us_states', con=tbl_engine).head()
```

# Store the us education CSV file to a dataframe

```python
# read the education table
excel_file = "resources/us_education.xls"
us_education_df = pd.read_excel(excel_file, sheet_name="Education 1970 to 2017")

us_education_df.head()
```

```python
# clean the education data
rep_us_education_df = us_education_df.drop([0, 1, 2, 4])
rep_us_education_df.reset_index(drop=True,inplace=True) 
# Create a new header variable from the first row of the dataframe
us_education_header = rep_us_education_df.iloc[0]
# Replace the dataframe with a new one which does not contain the first row
int_us_education_df = rep_us_education_df[1:]
# rebuild the header and reset the indexing
new_us_education_df =int_us_education_df.rename(columns = us_education_header).reset_index(drop=True)
# drop rows where the columns are null for column 2003 Rural-urban Continuum Code
new_us_education_df = new_us_education_df.dropna(subset=['2003 Rural-urban Continuum Code', '2003 Rural-urban Continuum Code'])
new_us_education_df.fillna(0,inplace=True)
# extract the state and count from fips code and create two new cols
new_us_education_df['fips_state_code'] = new_us_education_df['FIPS Code'].astype(str).str[:2]
new_us_education_df['fips_county_code'] = new_us_education_df['FIPS Code'].astype(str).str[2:]
new_us_education_df.head(1)
```

```python
# reformat the ecucation dataframe to the rows that will be loaded into the education table
new_us_education_df = new_us_education_df[["fips_state_code", 
                                           "fips_county_code", 
                                           "Less than a high school diploma, 2013-17",
                                           "High school diploma only, 2013-17",
                                           "Some college or associate's degree, 2013-17",
                                           "Bachelor's degree or higher, 2013-17",
                                          ]].copy()
new_us_education_df.rename(columns={"Less than a high school diploma, 2013-17":"hs_diploma_y_count",
                                    "High school diploma only, 2013-17":"hs_diploma_n_count",
                                    "Some college or associate's degree, 2013-17":"as_degree_count",
                                    "Bachelor's degree or higher, 2013-17":"ba_degree_count",
                                   },inplace=True)
new_us_education_df.head()
```

# Load the us education dataframe to the us_education MYSQL table

```python
# Create Engine and Pass in MySQL Connection
tbl_engine = create_engine("mysql://root:<amigo2598>@localhost/crime_db")
tbl_engine.table_names()
```

```python
# load the us states dataframe to the us states table
new_us_education_df.to_sql(name="us_education", con=tbl_engine, \
                         if_exists='append', index=False)
```

```python
# validate that the data was loaded into the table
pd.read_sql_query('select * from us_education', con=tbl_engine).head()
```
